 1. Substantiation of chosen sampling

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
```

Data preparation

```python
data = pd.read_csv('../dataset/typhoon_data.csv')

# Import label encoder from scikit-learn
from sklearn.preprocessing import LabelEncoder

sample = data.copy()

# Create a label encoder object
le = LabelEncoder()

# Define the mapping dictionary for typhoon grades to numeric values
grade_map = {'Tropical Depression': 0,
             'Tropical Storm': 1,
             'Severe Tropical Storm': 2,
             'Typhoon': 3,
             'Tropical Cyclone of TS intensity or higher': 4,
             'Just entering into the responsible area of RSMC Tokyo-Typhoon Center': 5,
             'Extra-tropical Cyclone': 6}

# Replace the typhoon grades with their corresponding numeric values
sample['grade'] = sample['grade'].replace(grade_map)

# Drop the first column (it is unnamed and holds index)
sample = sample.iloc[:, 2:]

# select only variables of interest and create a new dataset
sample_variables = ['year', 'month', 'day', 'Maximum sustained wind speed', 'The shortest radius of 30kt winds or greater', 'The longeast radius of 30kt winds or greater', 'grade', 'Central pressure']
sample = sample[sample_variables]

# Lets abbreviate
abbreviation = ['year', 'month', 'day', 'max_sust_wind_speed', 'shortest_rad_of_30kt_winds_or_greater', 'longest_rad_of_30kt_winds_or_greater', 'grade', 'central_pressure']
sample.columns = abbreviation

# Merge the year, month, and day columns and create a new column "date"
sample['date'] = pd.to_datetime(sample[['year', 'month', 'day']])

# # Drop the original year, month, and day columns
sample.drop(columns=['year', 'month', 'day'], inplace=True)

# Rearrange the columns so that the "date" column is first
sample = sample[['date', 'max_sust_wind_speed', 'shortest_rad_of_30kt_winds_or_greater', 'longest_rad_of_30kt_winds_or_greater', 'grade', 'central_pressure']]

sample = sample.set_index('date')
sample = sample.resample('M').mean() # aggregate average data by month

sample_monthly = sample.dropna()


sample_monthly
```

2. Stationary Analysis
The most common method to check whether a given data comes from a stationary series or not is simply plotting the data or some function of it

Raw visualization of our variables

```python
titles = ['Maximum sutained wind speed', 
        'Shortest Radius of 30kt', 
        'Longest Radius of 30kt', 
        'Typhoon Grade', 
        'Central Pressure']
        
feature_keys = ['max_sust_wind_speed', 'shortest_rad_of_30kt_winds_or_greater', 'longest_rad_of_30kt_winds_or_greater', 'grade', 'central_pressure']

colors = [ "purple", "orange", "green", "red", "pink", "olive"]

def show_raw_visualization(data):
    fig, axes = plt.subplots(
        nrows=3, ncols=2, figsize=(15, 10), dpi=80, facecolor="w", edgecolor="k"
    )
    for i in range(len(feature_keys)):
        key = feature_keys[i]
        c = colors[i % (len(colors))]
        t_data = data[key]
        t_data.index = data.index
        t_data.head()
        ax = t_data.plot(
            ax=axes[i // 2, i % 2],
            color=c,
            title="{}".format(titles[i]),
            rot=25,
        )
        ax.legend([titles[i]])
        ax.set(xlabel=None)
    plt.tight_layout()



show_raw_visualization(sample_monthly)

```

Stationarity

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

def stationarityRolling(df, col):
    # Run Augmented Dickey-Fuller test
    adf_result = sm.tsa.stattools.adfuller(df[col])
    
    # Plot rolling mean and standard deviation
    rolling_mean = df[col].rolling(window=12).mean()
    rolling_std = df[col].rolling(window=12).std()
    plt.plot(df.index, df[col], label='Original')
    plt.plot(df.index, rolling_mean, label='Rolling Mean')
    plt.plot(df.index, rolling_std, label='Rolling Std')
    plt.legend()
    plt.title(f'{col} Rolling Mean & Standard Deviation')
    plt.show()
    
    # Print ADF test results
    print(f'Augmented Dickey-Fuller Test Results for {col}:')
    print(f'ADF Statistic: {adf_result[0]}')
    print(f'p-value: {adf_result[1]}')
    print(f'Critical Values:')
    for key, value in adf_result[4].items():
        print(f'   {key}: {value}')
        
    # Determine if series is stationary based on p-value
    if adf_result[1] < 0.05:
        print(f'{col} is stationary')
    else:
        print(f'{col} is not stationary')
```

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

def stationarity(df, column_name):
    p_threshold = 0.05
    plt.figure(figsize=(8, 6))
    plt.plot(df.index, df[column_name])
    plt.xticks(rotation=45)
    plt.xlabel(u'Date', fontsize = 20) 
    plt.ylabel(column_name, fontsize = 20)
    plt.title(f'Time Series of {column_name}', fontsize = 20)
    
    # Testing for stationarity using Augmented Dickey-Fuller Test
    result = adfuller(df[column_name])
    print(column_name + ': Augmented Dickey-Fuller test:\n Statistic value = %5f\n p-value = %5f' % (result[0], result[1]))
    if result[1] < p_threshold:
        print(f'{column_name} variable is stationary\n')
    else:
        print(f'{column_name} variable is not stationary\n')
    
    # Decomposing the time series into trend, seasonal, and residual components
    decomposition = sm.tsa.seasonal_decompose(df[column_name], model='additive', period=12)
    trend = decomposition.trend
    seasonal = decomposition.seasonal
    residual = decomposition.resid
    
    # Plotting the decomposed components
    plt.figure(figsize=(8,6))
    plt.subplot(411)
    plt.plot(df[column_name], label='Original', color='blue')
    plt.legend(loc='best')
    plt.ylabel(column_name, fontsize = 12)
    plt.title(f"Time Series Decomposition of {column_name}", fontsize=15)
    plt.subplot(412)
    plt.plot(trend, label='Trend', color='blue')
    plt.legend(loc='best')
    plt.ylabel("Trend", fontsize = 12)
    plt.subplot(413)
    plt.plot(seasonal, label='Seasonality', color='blue')
    plt.legend(loc='best')
    plt.ylabel("Seasonality", fontsize = 12)
    plt.subplot(414)
    plt.plot(residual, label='Residuals', color='blue')
    plt.legend(loc='best')
    plt.ylabel("Residuals", fontsize = 12)
    plt.xlabel('Date', fontsize = 12)
    plt.tight_layout()
    plt.show()
```

```python
stationarityRolling(sample_monthly, 'max_sust_wind_speed')
stationarity(sample_monthly, 'max_sust_wind_speed')
```

```python
stationarityRolling(sample_monthly, 'shortest_rad_of_30kt_winds_or_greater')
stationarity(sample_monthly, 'shortest_rad_of_30kt_winds_or_greater')
```

```python
stationarityRolling(sample_monthly, 'longest_rad_of_30kt_winds_or_greater')
stationarity(sample_monthly, 'longest_rad_of_30kt_winds_or_greater')
```
