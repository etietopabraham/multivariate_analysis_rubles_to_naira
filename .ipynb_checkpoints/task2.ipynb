1. Plotting a non-parametric estimation of PDF in form of a histogram and kernel density function for MRV (or probability law in case of discrete MRV).

The goal of non-parametric estimation is to estimate the probability density function of a continuous random variable, or the probability mass function of a discrete random variable, without specifying the parameters of any particular probability distribution.

```python
def plotEstimation(dataset, column_names):
    figure, ax = plt.subplots(3, 2, figsize=(15, 15))
    sns.set_theme(style='whitegrid', palette="viridis")

    # Plot histograms or kernel density estimates for each column
    for i, column_name in enumerate(column_names):
        column_data = dataset[column_name]
        sns.histplot(column_data, kde=True, ax=ax.flatten()[i], stat='density')
        ax.flatten()[i].set_xlabel(column_name)
```

2. Estimation of multivariate mathematical expectation and variance.

```python
def estimation(dataset):
    # Compute the mean vector
    mean_vector = np.mean(dataset, axis=0)

    # Compute the variance vector
    var_vector = np.var(dataset, axis=0)

    # Create a table for mean and variance
    result_table = pd.DataFrame({'Mean': mean_vector, 'Variance': var_vector},
                                index=dataset.columns)

    # Print the table using tabulate
    print(tabulate(result_table, headers='keys', tablefmt='psql'))


# Compute estimations
estimation(sample)
```

Step 4. Estimation of pair correlation coefficients, their confidence intervals and significance levels.

```python
figure, ax = plt.subplots(1, 1, figsize=(10, 10))
sns.set_theme(style='whitegrid', palette='viridis')

corr_matrix = sample.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, ax=ax, vmax=1, fmt='.5g', linewidths=.5)

plt.title('Correlation')
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from tabulate import tabulate


# Compute the number of samples
n = len(sample)

# Compute the standard error of the Fisher z-transform
se = 1 / np.sqrt(n - 3)

# Compute the confidence intervals for each correlation coefficient
z_crit = 1.96  # 95% confidence interval
z_lower = np.arctanh(corr_matrix) - z_crit * se
z_upper = np.arctanh(corr_matrix) + z_crit * se
ci_lower = np.tanh(z_lower)
ci_upper = np.tanh(z_upper)

# Compute the p-values for each correlation coefficient
p_values = 2 * (1 - norm.cdf(np.abs(z_upper)))

# Create a table of the correlation matrix with confidence intervals and significance levels
table_data = []
for i in range(len(corr_matrix)):
    for j in range(i+1, len(corr_matrix)):
        if not np.isnan(corr_matrix.iloc[i,j]):
            pair = f"{corr_matrix.index[i]} - {corr_matrix.columns[j]}"
            corr = f"{corr_matrix.iloc[i,j]:.3f}"
            p_value = f"{p_values[i,j]:.3f}"
            ci = f"[{ci_lower.iloc[i,j]:.3f}, {ci_upper.iloc[i,j]:.3f}]"
            table_data.append([pair, corr, p_value, ci])

headers = ['Pair', 'Correlation', 'p-value', 'Confidence Interval']
print(tabulate(table_data, headers=headers))
```

6. Regression model,multicollinearity and regularization (if needed).

Ok, let's start by checking the variance inflation factor for each of our predictor variable

```python
# Import necessary libraries
from statsmodels.stats.outliers_influence import variance_inflation_factor
from tabulate import tabulate

# Create a dataframe with only the predictor variables
X = sample[['central_presure', 'max_sus_wind_speed', 'longest_rad_of_30kt_wind_or_greater', 'shortest_rad_of_30k_wind_or_greater']]

# Compute the VIF for each predictor variable
vif = pd.DataFrame()

# Create a list of tuples to hold the variable name and VIF value
vif_data = []
for i in range(len(X.columns)):
    vif = variance_inflation_factor(X.values, i)
    vif_data.append((X.columns[i], round(vif, 2)))

# Print the VIF values in a table
print(tabulate(vif_data, headers=['Variable', 'VIF'], tablefmt='psql'))
```

Before we perform ridge regression, we can also consider scaling our predictor variables to ensure that they have similar ranges. This can be done using the StandardScaler class from the scikit-learn library. Scaling the variables can help to improve the performance of ridge regression.

```python
from sklearn.preprocessing import StandardScaler

# Scale the predictor variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Check the mean and standard deviation of the scaled variables
print(np.mean(X_scaled, axis=0))
print(np.std(X_scaled, axis=0))
```

Now let's ridge regress

```python
# Import the Ridge regression model
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Define the predictor and response variables
X = graded_sample[['central_presure', 'max_sus_wind_speed', 'longest_rad_of_30kt_wind_or_greater', 'shortest_rad_of_30k_wind_or_greater']].values
y = graded_sample['grade'].values

# Scale the predictor variables
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fit the Ridge regression model on the training set
ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)

# Make predictions on the test set
y_pred = ridge.predict(X_test)

# Evaluate the model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean squared error: ", mse)
print("R-squared: ", r2)

```

That didn't go well, so let's try lasso regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the predictor variables
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Fit the Lasso regression model to the training data
lasso = Lasso(alpha=0.1, max_iter=10000)
lasso.fit(X_train_scaled, y_train)

# Make predictions using the testing set
y_pred = lasso.predict(X_test_scaled)

# Calculate and print the mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error: ", mse)

# Calculate and print the R-squared value
r2 = r2_score(y_test, y_pred)  
print("R-squared: ", r2)
```

But we can also try Elastic Net regression, a combination of both Ridge and Lasso regression, which means it can handle multicollinearity and feature selection simultaneously. It adds both L1 (Lasso) and L2 (Ridge) penalties to the loss function to regularize the coefficients.

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error, r2_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Fit the model with Elastic Net regularization
enet = ElasticNet(alpha=0.1, l1_ratio=0.5)
enet.fit(X_train, y_train)

# Predict the response for the test dataset
y_pred = enet.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean squared error: ", mse)
print("R-squared: ", r2)
```
